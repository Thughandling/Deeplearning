### 1.  3개의 각으로 삼각형의 예각, 직각, 둔각을 구별하는 프로그램을 만들어라.

```python
def  angle(t):
    result = ' '
    if 90 in t and min(t) > 0 :
        result ='직각삼각형'
    elif max(t) > 90 :
       
        result = '둔각삼각형'
    elif min(t) ==0 or sum(t) <180 or sum(t) > 180 or len(t)>3:
        result = '삼각형이 아니다'
    else:
        result = "예각삼각형"
    return result
X = [[60, 60, 60],[30, 60, 90],[20, 40, 120],[0, 90, 90],[60, 70, 80],[40, 40, 50, 50]]
for i in X:
    print(i,"=",angle(i))
    
    
```

### 3.  크롬드라이버 및 셀레니움을 사용하여 벅스,멜론, (지니)의 인기차트 데이터를 각각  50위까지 추출한 후,
모두 통합하여 데이터프레임으로 저장하고 엑셀로 저장하시오.


```python
from selenium import webdriver
import time
from bs4 import BeautifulSoup
```

```python
bugs = webdriver.Chrome('chromedriver.exe')
bugs.get('https://music.bugs.co.kr/chart')
bugs_lank = []
```

```python
melon = webdriver.Chrome('chromedriver.exe')
melon.get('https://www.melon.com/chart/index.htm')
melon_lank = []
```

```python
genie = webdriver.Chrome('chromedriver.exe')
genie.get('https://www.genie.co.kr/chart/top200')
genie_lank = []
```

```python
for i in range(50):
    bugs_lank.append(bugs.find_elements_by_css_selector('p.title')[i].text)
    melon_lank.append(melon.find_elements_by_css_selector('div.wrap_song_info > .rank01')[i].text)
    genie_lank.append(genie.find_elements_by_css_selector('a.title.ellipsis')[i].text)
```

```python
print(len(bugs_lank),len(melon_lank),len(genie_lank))
import pandas as pd
data = {'bugs':bugs_lank, 'melon':melon_lank,'genie':genie_lank }
```

```python
Data = pd.DataFrame(data)
Data.index = Data.index+1
Data
```
