[2] 낙차가 가장 큰상자를 구하라.낙차의 높이를 출력하는것이 문제

데이터 접근 : 각 박스 높이 데이터를, 인덱스를 통해 접근 => enumerate 사용.

낙차 계산: 
결론적으로 맨왼쪽 박스가 오른쪽으로 붙는 상황, 값의 차이를 통해 해결하고자함.

```python
gravity = [7,4,2,0,0,6,0,7,0] 
grab = enumerate(gravity)
count = 0 

result= range(len(gravity))
result = list(result)
result[-1]=0 #마지막은 떨어질수 없으니 0으로 초기화
End =max(range(len(gravity)))#최하층(땅)

type(result)
```

배경 : 가장 높은 위치의 상자가 당연히 낙차가 큰 상황임.

(0,7) 즉, 첫 인덱스이며 7 높이의 데이터를 한칸 씩이동하며 높이를 비교하는데,
비교 데이터가 값이 같거나 높으면, 그 순간에 낙차가 발생하는 것임.

데이터를 이동하면서, 통과 할 때 마다 count 변수를 통해 이동한 횟수(낙차 범위)를 셈하고,

for문 내에서, i가 0일때의 비교된 카운트 값을, 결과 List인 result에 일치하는 인덱스에 count 값을 저장한다.

저장된 값을 통해 max를 뽑아낸다.

```python
for i, v in grab: #인덱스에 해당하는 높이 분리 사용
    count = 0
    
    for j in range(i+1,len(gravity)): #인덱스 다음 데이터부터 비교 i+1
           
            if j==End : #비교 대상의 위치가 지면일 경우, 관측 값은 지면 바로 1칸 위라는 듯
                count+=1
                result[i] =count
                continue
            elif gravity[i]==0: #움직일 상자가 없는 경우 낙차를 0으로
                result[i] = 0
                break
            
            elif gravity[i] > gravity[j]:
                count+=1
                
            elif gravity[i] <= gravity[j]:
                count+=1
                #print(gravity[i],gravity[j])
                #print(i, count)
                
                result[i]= count
                break
Max = max(result)
for i in range(len(result)):
    if result[i] == Max:
        print(i,"번째 박스 최대 낙차 => ",Max, "\n인덱스 해당 낙차 값 => ",result)
```

[4] while문을 사용하여 다음과 같이 별(*)을 표시하는 프로그램을 작성해 보자.


```python
#while 사용
i=1
v=4 

while i<6:
    print(" "*v +"*"*i)
    v-=1
    i+=1
    
```
